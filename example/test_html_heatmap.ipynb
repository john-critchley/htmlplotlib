```python
from collections import deque

def colliexcel(n):
    dq=deque()
    while n:
        n,r=divmod(n,26)
        dq.appendleft(chr(r+ord('A')-1))
    return str().join(dq)


import os
import sys
from pathlib import Path
import numpy as np

os.path.join(Path.home(),'src/htmlplotlib')

#print(sys.path)
#print(os.environ['PYTHONPATH'])
sys.path.insert(0, os.path.join(Path.home(),'src','htmlplotlib'))
import htmlplotlib

import htmlplotlib

import numpy as np
import htmlplotlib

sx = sy = 12
hor = False

# Ensure data is the correct size
if 'data' not in vars() or data.shape!=(sx,sy) or True:
    data = np.random.rand(sx, sy)

# Generate labels if not provided
xticklabels = np.vectorize(colliexcel)(np.arange(1, sx+1))
yticklabels = np.arange(1, sy+1)
```

```python
import pandas as pd
```

```python
data = np.random.rand(sx, sy)*1.2+0.25
pd.DataFrame(data, index=yticklabels, columns=xticklabels)
```

```python
hor=False
htmlplotlib.html_heatmap(data, cmap='rocket', show=True, xlabel='X Axis', ylabel='Y Axis',
                         xticklabels=xticklabels, yticklabels=yticklabels, square=True,     cbar_kws={'orientation': 'horizontal' if hor else 'vertical'})
```
