```python
from collections import deque
import os
import sys
from pathlib import Path
import numpy as np
```

```python
def colliexcel(col_num):

    dq = deque()
    col_num-=1
    while col_num >= 0:
        col_num, remainder = divmod(col_num, 26)
        dq.appendleft(chr(remainder + ord('A')))
        col_num -= 1
    return str().join(dq)

# Testing the function with the first 5 columns

print(*(colliexcel(i)for i in range(1,99)))

os.path.join(Path.home(),'src/htmlplotlib')

#print(sys.path)
#print(os.environ['PYTHONPATH'])
sys.path.insert(0, os.path.join(Path.home(),'src','htmlplotlib'))

import numpy as np
import pandas as pd
import htmlplotlib

sx = 10; sy = 20
hor = False

# Ensure data is the correct size
if 'data' not in vars() or data.shape!=(sx,sy) or True:
    data = np.random.rand(sx, sy)

# Generate labels if not provided
xticklabels = np.vectorize(colliexcel)(np.arange(1, sy+1))
yticklabels = np.arange(1, sx+1)
```

```python
#data = np.random.rand(sx, sy)*1.2+0.25
data = np.random.rand(sx, sy)*100

df=pd.DataFrame(data, index=yticklabels, columns=xticklabels)
df
```
